```python
import re

def main():
    # ------------------------------
    # 1. CSV-Datei einlesen
    # ------------------------------
    with open('./orders.csv', 'r', encoding='utf-8') as f_in:
        text = f_in.read()

    # ------------------------------
    # 2. Alle Order Numbers extrahieren (bereits funktioniert)
    # ------------------------------
    order_numbers = re.findall(r'\d+', text)
    print("Order Numbers:", order_numbers)

    # ------------------------------
    # 3. Alle Product Codes extrahieren
    # Annahme: Codes sind z.B. 3 Buchstaben + 3 Zahlen (ABC123)
    # ------------------------------
    product_codes = re.findall(r'[A-Z]{3}\d{3}', text)
    print("Product Codes:", product_codes)

    # ------------------------------
    # 4. Alle Preise extrahieren
    # Annahme: Preise im Format 123 oder 123.45
    # ------------------------------
    prices = re.findall(r'\d+(?:\.\d+)?', text)
    prices = [float(p) for p in prices]
    print("Prices:", prices)

    # ------------------------------
    # 5. Alle Order Dates extrahieren
    # Annahme: Datum im Format YYYY-MM-DD
    # ------------------------------
    dates = re.findall(r'\d{4}-\d{2}-\d{2}', text)
    print("Order Dates:", dates)

    # ------------------------------
    # 6. Orders ber $500 finden
    # ------------------------------
    orders_over_500 = [float(p) for p in prices if float(p) > 500]
    print("Orders over $500:", orders_over_500)

    # ------------------------------
    # 7. Datum ins Format DD/MM/YYYY ndern
    # ------------------------------
    formatted_dates = [re.sub(r'(\d{4})-(\d{2})-(\d{2})', r'\3/\2/\1', date) for date in dates]
    print("Formatted Dates:", formatted_dates)

    # ------------------------------
    # 8. Orders mit hchster Menge
    # Annahme: Menge ist letzte Zahl jeder Zeile
    # ------------------------------
    quantities = []
    # Zeilen splitten, Header berspringen
    lines = text.strip().split('\n')[1:]
    for line in lines:
        parts = re.findall(r'\d+', line)
        if parts:
            quantities.append(int(parts[-1]))  # letzte Zahl = Menge
    if quantities:
        max_quantity = max(quantities)
        print("Max Quantity:", max_quantity)
    else:
        max_quantity = 0
        print("Keine Mengen gefunden")

    # ------------------------------
    # 9. Gnstigste Orders
    # ------------------------------
    if prices:
        min_price = min(prices)
        print("Cheapest Price:", min_price)
    else:
        print("Keine Preise gefunden")

if __name__ == '__main__':
    main()
```
